```python
df1 = df[['Type',"Time"]]
```

```python
df_dragon = df1.query('Type=="DRAGON"')
df_nashor = df1.query('Type=="BARON_NASHOR"')
df_herald = df1.query('Type=="RIFT_HERALD"')

df_dragonE_feu = df1.query('Type=="FIRE_DRAGON"')
df_dragonE_eau = df1.query('Type=="WATER_DRAGON"')
df_dragonE_terre = df1.query('Type=="EARTH_DRAGON"')
df_dragonE_air = df1.query('Type=="AIR_DRAGON"')
df_dragonE_vieu = df1.query('Type=="ELDER_DRAGON"')
```

```python


fig, ax = plt.subplots(figsize=(20,15))
plt.title("Histograme de l'élimination des nashor dans le temps", fontsize=20)
plt.xlabel("Temp", fontsize=15)
plt.ylabel("Nombre de créature tuer", fontsize=15)

plt.hist([df_nashor.Time, df_herald.Time], stacked=True,bins=100)
plt.legend(["Baron Nashor","Rift Herald"])
plt.savefig("../../../nashor.png")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20,15))
plt.title("Histograme des élimination de dragon élémentaire dans le temps", fontsize=20)
plt.xlabel("Temp", fontsize=15)
plt.ylabel("Nombre de créature tuer", fontsize=15)

plt.hist([df_dragonE_feu.Time, df_dragonE_eau.Time, df_dragonE_terre.Time, df_dragonE_air.Time, df_dragonE_vieu.Time],
         stacked=True,color=["#ffc0c0", "#9BB7D4", "#BCDEBB","#53B0AE", "#C74375"], bins=100)

plt.legend(["Dragon Feu","Dragon Eau","Dragon Terre", "Dragon Aire", "Dragon Ancien"])
plt.show()
```

```python
df_kills = pd.read_csv("../../../Data/DataRaw/set_kills.csv")
df_kills = df_kills.drop(columns="Unnamed: 0")

df_kills_herald = df1[df1.Type =="FIRE_DRAGON"]
tt_tepms5 = df_kills_herald.copy()
tt_tepms5.Time = tt_tepms5.Time.round(0)
tt_tepms5 = pd.pivot_table(tt_tepms5, index="Time",aggfunc="count")
tt_tepms5 = tt_tepms5.reset_index()

df_kills_herald = df1[df1.Type =="EARTH_DRAGON"]
tt_tepms4 = df_kills_herald.copy()
tt_tepms4.Time = tt_tepms4.Time.round(0)
tt_tepms4 = pd.pivot_table(tt_tepms4, index="Time",aggfunc="count")
tt_tepms4 = tt_tepms4.reset_index()

df_kills_herald = df1[df1.Type =="WATER_DRAGON"]
tt_tepms3 = df_kills_herald.copy()
tt_tepms3.Time = tt_tepms3.Time.round(0)
tt_tepms3 = pd.pivot_table(tt_tepms3, index="Time",aggfunc="count")
tt_tepms3 = tt_tepms3.reset_index()

df_kills_herald = df1[df1.Type == "AIR_DRAGON"]
tt_tepms2 = df_kills_herald.copy()
tt_tepms2.Time = tt_tepms2.Time.round(0)
tt_tepms2 = pd.pivot_table(tt_tepms2, index="Time",aggfunc="count")
tt_tepms2 = tt_tepms2.reset_index()

df_kills_herald = df_kills
tt_tepms6 = df_kills_herald.copy()
tt_tepms6.Time = tt_tepms6.Time.round(0)
tt_tepms6 = pd.pivot_table(tt_tepms6, index="Time",aggfunc="count")
tt_tepms6 = tt_tepms6.reset_index()
tt_tepms6 = tt_tepms6.drop(columns=["Assist_1","Assist_2","Assist_3","Assist_4","Date","Killer","Team","Victim","x_pos","y_pos"])


df_kills_herald = df1[df1.Type == "ELDER_DRAGON"]
tt_tepms1 = df_kills_herald.copy()
tt_tepms1.Time = tt_tepms1.Time.round(0)
tt_tepms1 = pd.pivot_table(tt_tepms1, index="Time",aggfunc="count")
tt_tepms1 = tt_tepms1.reset_index()
tt_tepms1 = tt_tepms1[tt_tepms1.Time < 70]

fig, ax = plt.subplots(figsize=(20,10))

plt.ylabel("Nombres de tuer")
tt_tepms5.plot(x="Time",y="Type",marker='.', label="Fire Dragon", color="red",ax =ax)
tt_tepms4.plot(x="Time",y="Type",marker='.', label="Earth dragon", color="green",ax =ax)
tt_tepms3.plot(x="Time",y="Type",marker='.', label="Water Dragon", color="blue",ax =ax)
tt_tepms2.plot(x="Time",y="Type",marker='.', label="Air Dragon", color="orange",ax =ax)

tt_tepms1.plot(x="Time",y="Type",marker='.', label="Elder Dragon", color="purple",ax =ax)

plt.xlabel("Temps (en jeu)")
plt.xticks(np.arange(0, int(df_kills.Time.max())+1,2))
plt.xlim([tt_tepms5.Time.min(),tt_tepms1.Time.max()])




plt.show()
```
