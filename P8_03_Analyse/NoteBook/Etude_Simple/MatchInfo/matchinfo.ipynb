```python

fig, ax = plt.subplots(figsize=(15,10))

df.Type.value_counts().plot.bar()

plt.title("RÃ©partition des differents type de partie", fontsize=20)
plt.xlabel('Type de matchs',fontsize=15)
plt.xticks(rotation=0)
plt.ylabel('Nombres de match',fontsize=15)
plt.savefig("../../../ImgGraph/repartitionType.png")
plt.show()
```

```python

fig, ax = plt.subplots(figsize=(15,10))

df.League.value_counts().plot.bar()

plt.title("RÃ©partition des differents type de league", fontsize=20)
plt.xlabel('Type de league',fontsize=15)
plt.xticks(rotation=0)
plt.ylabel('Nombres de match',fontsize=15)
plt.savefig("../../../ImgGraph/repartitionLeague.png")
plt.show()
```

```python

test = df.pivot_table(index="Year", values="bResult",aggfunc='count')
test = test.reset_index()
test.columns = ["Year","Nb_match"]


test.plot.bar(x="Year", y="Nb_match", figsize=(15,10))

plt.title("RÃ©partition du nombre de match par ans", fontsize=20)
plt.xlabel('AnnÃ©e des matchs',fontsize=15)
plt.xticks(rotation=0)
plt.ylabel('Nombres de match',fontsize=15)
plt.savefig("../../../ImgGraph/repartitionAns.png")
plt.show()
```
