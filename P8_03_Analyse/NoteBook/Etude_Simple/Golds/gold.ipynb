```python
## Try todo, prendre chaque partie et en faire une liste pour des graph ?

a = pd.DataFrame()
i=0
plt.figure(figsize=(20,15))
a

while i <= 3000:
    
    df_tempo = df_total[df_total.IdMatch == i]
    
    compt = 0
    j = 5
    while j <= 99:
        
        if ~np.isnan(df_tempo.iloc[0,j]): 
            compt+=1
            a.loc[j,"Heure"] = compt
            a.loc[j,"Or"] = df_tempo.iloc[0,j]
         
        j+=1
    plt.plot(a.Heure,a.Or,'o',color='green')

    i+=1
plt.xlabel("Temps dans la partie")
plt.ylabel("Nombre de pieces Or")
plt.show()

```

```python
for i in [0]:
    bluewin = df_total.query("Type =='goldblue' & MatchWinner == 'Blue'").IdMatch
    redwin = df_total.query("Type =='goldred' & MatchWinner == 'Red'").IdMatch

    bluelose = df_total.query("Type =='goldblue' & MatchWinner == 'Red'").IdMatch
    redlose = df_total.query("Type =='goldred' & MatchWinner == 'Blue'").IdMatch

    b = df_total[df_total.IdMatch.isin(bluewin.append(redwin))]
    b = b.reset_index()
    b = b.drop(columns="index")
    
    b1 = df_total[df_total.IdMatch.isin(bluelose.append(redlose))]
    b1 = b1.reset_index()
    b1 = b1.drop(columns="index")
    
    a = pd.DataFrame()
    
    
    i=0
    plt.figure(figsize=(20,15))
    a

    while i <= 3000:

        df_tempo = b.iloc[i]
        df_tempo2 = b1.iloc[i]

        compt = 0
        j = 5
        while j <= 99:

            if ~np.isnan(df_tempo.iloc[j]): 
                compt+=1
                a.loc[j,"Heure"] = compt
                a.loc[j,"Or"] = df_tempo.iloc[j]
            
            if ~np.isnan(df_tempo2.iloc[j]): 
                compt+=1
                a.loc[j,"Heure"] = compt
                a.loc[j,"OrPerdant"] = df_tempo2.iloc[j]


            j+=1
        plt.plot(a.Heure,a.Or,'r', alpha=.1)
        plt.plot(a.Heure,a.OrPerdant,'b', alpha=.1)

        i+=1

plt.show()

```
